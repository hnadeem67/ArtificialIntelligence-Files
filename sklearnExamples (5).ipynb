```python
import seaborn as sns
import pandas as pd
import numpy as np

# sklearn imports below
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.datasets import load_iris
```

```python
def mlBayes(X, y, p):
  # init y_predict
  y_pred = -1

  nbObj = GaussianNB()   # create an instance of the object
  nbObj.fit(X,y)        # fit or train the model
  y_pred = nbObj.predict(p)

  return y_pred

def mlSVM(X, y, p):
  # init y_predict
  y_pred = -1
  
  # fit or train the model
  svcObj = SVC()
  svcObj.fit(X,y)
  y_pred = svcObj.predict(p)

  return y_pred

def mlRegression(X, y, p):
  # init y_predict
  y_pred = -1
  
  regObj = LinearRegression()
  regObj.fit(X, y)
  y_pred = regObj.predict(p)

  return y_pred

def mlRandomForest(X, y, p):
  # init y_predict
  y_pred = -1

  rfObj = RandomForestClassifier()
  rfObj.fit(X, y)
  y_pred = rfObj.predict(p)

  return y_pred
```

```python
def main():
  Features = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
  Class = np.array([1, 1, 1, 2, 2, 2])
  Predict = [[-0.8, -1]]

  # download Iris Data 
  Features, Class = load_iris(return_X_y=True)
  Predict = [[-0.8, -1, 0, 0]]
  print(Features)
  print(Class)

  y_bayes = mlBayes(Features, Class, Predict)
  y_svm = mlSVM(Features, Class, Predict)
  y_reg = mlRegression(Features, Class, Predict)
  y_rf  = mlRandomForest(Features, Class, Predict)

  print('The predicted value is bayes = ', y_bayes)
  print('The predicted value is svm   = ', y_svm)
  print('The predicted value is reg   = ', y_reg)
  print('The predicted value is rfores= ', y_rf)

  return 
```

```python
main()
```
