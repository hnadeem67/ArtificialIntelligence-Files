HYPOTHESIS: Segregate Burrows based on type of crime. 

```python
#print(df1.sort_values(by=['LAW_CAT_CD'=='FELONY']))
```

```python
df1['FELONY'] = pd.Categorical(df1['LAW_CAT_CD'],categories=['FELONY'])
df1['FELONY']
```

```python
# MISDEMEANOR in different burrows ('BRONX', 'QUEENS', 'MANHATTAN',
#                                     'STATEN ISLAND','BROOKLYN')

df1['Mis'] = pd.Categorical(df1['LAW_CAT_CD'],categories=['MISDEMEANOR'])
lengthMis = len(df1['Mis']) - df1['Mis'].isnull().sum()
print(lengthMis)
# Brooklyn
df1['MisBrooklyn'] = pd.Categorical(df1['BORO_NM'],categories=['BROOKLYN'])
lengthMisBrook = len(df1['MisBrooklyn']) - df1['MisBrooklyn'].isnull().sum()
print('Number of Misdemnors in Brooklyn',lengthMisBrook)
#Bronx
df1['MisBronx'] = pd.Categorical(df1['BORO_NM'],categories=['BRONX'])
lengthMisBronx = len(df1['MisBronx']) - df1['MisBronx'].isnull().sum()
print('Number of Misdemnors in Bronx',lengthMisBronx)
# QUEENS
df1['MisQueens'] = pd.Categorical(df1['BORO_NM'],categories=['QUEENS'])
lengthMisQueens = len(df1['MisQueens']) - df1['MisQueens'].isnull().sum()
print('Number of Misdemnors in Queens',lengthMisQueens)
#MANHATTAN
df1['MisMan'] = pd.Categorical(df1['BORO_NM'],categories=['MANHATTAN'])
lengthMisMan = len(df1['MisMan']) - df1['MisMan'].isnull().sum()
print('Number of Misdemnors in Manhattan',lengthMisMan)
#STATEN ISLAND
df1['MisSI'] = pd.Categorical(df1['BORO_NM'],categories=['STATEN ISLAND'])
lengthMisSI = len(df1['MisSI']) - df1['MisSI'].isnull().sum()
print('Number of Misdemnors in Staten ISland',lengthMisSI)
```
