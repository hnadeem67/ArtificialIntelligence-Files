```python
import numpy as np
import pandas as pd 
import seaborn as sns

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
def mkData():

  # y = 2x - 1
  # y = 2x**2 + 2x - 1

  #x = [-1, 0, 1, 2, 3, 4]
  #y = [-3, -1, 1, 3, 5, 7]

  start = -1
  stop  = 10
  count = 1

  x = np.arange(start, stop, count)
  y = []

  for n in x:
    v = (2*n)**2 + (2*n) -1
    y.append(v)

  return np.array(x), np.array(y)

def mkChart(x,y):
  sns.scatterplot(x=x, y=y)
  return 0
```

```python
def main():
  # generate some data
  xs, ys = mkData()
  status = mkChart(xs, ys)

  # create the model, fit, predict
  model = Sequential((Dense(units=1, input_shape=[1])))
  model.compile(optimizer='sgd', loss='mean_squared_error')
  model.fit(xs, ys, epochs=300)
  print(model.predict([10.0]))

  return 0
```

```python
# run main
r = main()
```
